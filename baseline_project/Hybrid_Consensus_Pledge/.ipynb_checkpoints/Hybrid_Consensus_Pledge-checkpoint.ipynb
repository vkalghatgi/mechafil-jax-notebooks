## Hybrid Consensus Pledge 

The document [here](https://hackmd.io/haBzt9PSQLSsfaFQr4Sijg?view) introduces a hybrid consensus pledge model analagous to the hybrid minting model in order to mitigate the risks present in the current initial pledge model as dicussed in the introduction to the baseline pledge "bug" [document](https://docs.google.com/document/d/1V3Hm1uKemnuVBmdW3KJYqeecA5-A_qf2eQana-9NuUA/edit#heading=h.qa12l2lw9n0)

- The proposal introduces Hybrid Consensus Pledge model analagous to the Hybrid Minting model in order to mitigate the risks of the current initial pledge model such that consensus pledge is split in to a "simple" and "baseline" component:
    - A sector's simple consensus pledge is solely dependant on the sectors QAP with respect to the network's upon sector commit. 
    - A sectors baseline consensus pledge will maintain the current initial pledge property in which it is divided by the maximum of Network QAP and the baseline $b(t)$. 
- Similarly to how simple minting provides a floor on network rewards, simple consensus pledge can provide a reasonable floor on collateral requirements whilst also maintaining onboarding incentives through exponentially decreasing baseline consensus `pledge.

The Current Pledge Specification is shown below: 

- $P$: Initial Pledge 
- $P_{s}$: Initial Storage Pledge
- $P_{c}$: Initial Consensus Pledge 
- $Q$: Network QAP 
- $q$: Sector QAP
- $b(t)$: Filecoin's exponential baseline function
- $S$: Filecoin Circulating Supply 
- $\alpha$: Target lock (currently: $\alpha = 0.3$)

Filecoin initial pledge for a sector committed at time $t$ with QAP $q$ is as follows: 

$$ P(t) = P_{S}(t) + P_{C}(t) $$

where the initial pledge for sector with QAP $q$ committed at time $t$ is 

$$ P_{S}(t) = E[20DaysBlockReward] $$

and

$$P_{C}(t) = \alpha S(t) \cdot \frac{q}{max(Q(t), b(t))} $$

#### Proposal: 
$$P_C(t) = (1-\gamma')P_{C,S}(t) + \gamma' P_{C,B}(t) $$

Initially, set $\gamma' = 0.7$: (Note, $\gamma$ can take on any value such that $0 \le \gamma' \le 1$, but start with $\gamma' = \gamma$ per the minting model's specification:  

$$ P_{C,S}(t) = \alpha S(t) \cdot \frac{q}{Q(t)} $$
$$ P_{C,B}(t) = \alpha S(t) \cdot \frac{q}{max(Q(t), b(t))} $$

**The crucial innovation here is that there is a "simple consensus" portion of pledge that will always be determined by a sectors QAP *in relation to the Network's QAP* and then there is a "baseline consensus" portion to pledge that will have the exponentially decreasing property as is in the current protocol specification (in which pledge is normalized by** $max(NetworkQAP(t), b(t))$

## Simulations 

### Current Protocol Spec: $\gamma' = 1$ 

```python
def add_generated_quantities(cil_rbp, duration=365)->pd.DataFrame:
    # add ROI to trajectory df
    cil_rbp['day_pledge_per_QAP'] = SECTOR_SIZE * (cil_rbp['day_locked_pledge']-cil_rbp['day_renewed_pledge'])/(cil_rbp['qa_day_onboarded_power_pib']*PIB)
    cil_rbp['day_rewards_per_sector'] = SECTOR_SIZE * cil_rbp.day_network_reward / (cil_rbp.network_QAP_EIB * EIB)
    
    cil_rbp['1y_return_per_sector'] = cil_rbp['day_rewards_per_sector'].rolling(365).sum().shift(-365+1).values.flatten()
    cil_rbp['1y_sector_roi'] = cil_rbp['1y_return_per_sector'] / cil_rbp['day_pledge_per_QAP']

    return cil_rbp

```

```python
forecast_length
```

```python
# JAX Simulation Per Current protocol spec
gamma = 1
gamma_weight_type = 1
lock_target_val = 0.3
sector_duration = 365
scenario2results = {}
gamma_1_scenario_name = 'Gamma 1'
scenario2results[gamma_1_scenario_name] = {}
lock_target_trajectory = jnp.ones(forecast_length) * lock_target_val
for scenario_name, trajectories in scenarios.items():
    
    rbp = jnp.array(trajectories['rbp'])
    rr = jnp.array(trajectories['rr'])
    fpr = jnp.array(trajectories['fpr'])
    
    simulation_results = sim.run_sim(
        rbp,
        rr,
        fpr,
        lock_target_trajectory,
        start_date,
        current_date,
        forecast_length,
        sector_duration,
        offline_data, 
        gamma, 
        gamma_weight_type, 
    )
    
    simulation_results = pd.DataFrame(simulation_results)
    simulation_results = add_generated_quantities(simulation_results, duration=365)
    scenario2results[gamma_1_scenario_name][scenario_name] = simulation_results
```

### Hybrid Consensus Pledge Model ($\gamma' = 0.7$) 

```python
gamma_0_7_scenario_name = 'Gamma 0.7'
scenario2results[gamma_0_7_scenario_name] = {}
gamma = 0.7
lock_target_val = 0.3
gamma_weight_type = 1

for scenario_name, trajectories in scenarios.items():
    rbp = jnp.array(trajectories['rbp'])
    rr = jnp.array(trajectories['rr'])
    fpr = jnp.array(trajectories['fpr'])
    lock_target_trajectory = jnp.ones(forecast_length) * lock_target_val

    simulation_results = sim.run_sim(
        rbp,
        rr,
        fpr,
        lock_target_trajectory,
        start_date,
        current_date,
        forecast_length,
        sector_duration,
        offline_data,
        gamma,  # Gamma = 0.7
        gamma_weight_type, 
    )

    simulation_results = pd.DataFrame(simulation_results)
    simulation_results = add_generated_quantities(simulation_results, duration=365)
    
    scenario2results[gamma_0_7_scenario_name][scenario_name] = simulation_results
```

### Simulation Plots 

```python
def plot_x(ax, results, key, labels, colors, scale_by=1, linestyle='-'):
    for ii, scenario in enumerate(labels):
        res = results[scenario]
        ax.plot(t, res[key]/scale_by, label=scenario, color=colors[ii], linestyle=linestyle)
        
    for tick in ax.get_xticklabels():
        tick.set_rotation(60)
```

```python
total_sim_len = (end_date - start_date).days
t = [start_date + timedelta(days=i) for i in range(total_sim_len)]

fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(10,10))

c_vec = [c(0.6) for c in colors]

for gamma_scenario in scenario2results.keys():
    if gamma_scenario == gamma_0_7_scenario_name: 
        linestyle = '--'
    else: 
        linestyle = '-'
    keys = list(scenario2results[gamma_scenario].keys())
    axx = ax[0, 0]
    plot_x(axx, scenario2results[gamma_scenario], 'rb_total_power_eib', keys, c_vec, linestyle=linestyle)
    axx.set_yscale('log')
    axx.set_ylabel('EiB')
    axx.set_title('RBP')
    # Dont need to duplicate legends

    axx = ax[0, 1]
    plot_x(axx, scenario2results[gamma_scenario], 'qa_total_power_eib', keys, c_vec, linestyle=linestyle)
    axx.set_yscale('log')
    axx.set_ylabel('EiB')
    axx.set_title('QAP')
    if gamma_scenario == gamma_1_scenario_name: 
        axx.legend(bbox_to_anchor=(1,0.5))

    axx = ax[1, 0]
    plot_x(axx, scenario2results[gamma_scenario], 'day_network_reward', keys, c_vec, linestyle=linestyle)
    axx.set_ylabel('FIL/day')
    axx.set_title('Minting Rate')

    axx = ax[1, 1]
    plot_x(axx, scenario2results[gamma_scenario], 'network_locked', keys, c_vec, scale_by=1e6, linestyle=linestyle)
    axx.set_ylabel('M-FIL')
    axx.set_title('Network Locked')
    axx.axhline(0, linestyle=':', color='k')

    axx = ax[2, 0]
    plot_x(axx, scenario2results[gamma_scenario], 'circ_supply', keys, c_vec, scale_by=1e6, linestyle=linestyle)
    axx.set_ylabel('M-FIL')
    axx.set_title('Circulating Supply')

    axx = ax[2, 1]

    for ii, scenario in enumerate(keys):
        res = scenario2results[gamma_scenario][scenario]
        axx.plot(t, res['network_locked']/res['circ_supply'], label= str(gamma_scenario)+':'+keys[ii], color=c_vec[ii], linestyle=linestyle)
    for tick in axx.get_xticklabels():
        tick.set_rotation(60)
    axx.set_title('L/CS')
    
    #axx = ax[3,0]
    #plot_x(axx, scenario2results[gamma_scenario], 'day_pledge_per_QAP', keys, c_vec, linestyle=linestyle)
    #axx.set_ylabel('FIL')
    #axx.set_title('Pledge/32GiB QA Sector')
    #axx.set_ylim(0)
    
   #axx = ax[3,1]
    #plot_x(axx, scenario2results[gamma_scenario], '1y_sector_roi', keys, c_vec, scale_by=.01, linestyle=linestyle)
    #axx.set_ylabel('FIL')
    #axx.set_title('1Y Realized FoFR')
    #axx.set_ylim(0)

fig.suptitle('Network Macroeconomic KPIs: Current vs Hybrid Pledge Model')
legend_handles = [
    Line2D([], [], linestyle='-', color='black', label='Current Pledge Spec: (Gamma=1)'),
    Line2D([], [], linestyle='--', color='black', label='Hybrid Consensus Pledge (Gamma = 0.7)'),
]

# Add the legend to the plot
fig.legend(handles=legend_handles, bbox_to_anchor=(1, 0.7), loc='center right')
plt.tight_layout()

plt.savefig('Macroeconomic_KPIs:_Hybrid_Consensus_Pledge.png', bbox_inches='tight')

# Adjust the layout to make space for the legend
#plt.subplots_adjust(right=0.85)

# Display the plot
#plt.show()
```
